# International Debt Analysis System - Exploratory Data Analysis (EDA)

This notebook performs EDA on the cleaned international debt dataset using Pandas, Matplotlib, Seaborn, and Plotly.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

## 1. Load Data

```python
df = pd.read_csv('../data/cleaned/cleaned_debt_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## 2. Country-wise Debt Distribution

```python
country_debt = df.groupby('country_name')['debt_value'].sum().reset_index().sort_values(by='debt_value', ascending=False)

fig = px.bar(
    country_debt.head(15), 
    x='country_name', 
    y='debt_value', 
    title='Top 15 Countries with Highest Total Debt',
    labels={'country_name': 'Country', 'debt_value': 'Total Debt (USD)'}
)
fig.show()
```

## 3. Indicator-wise Debt Distribution

```python
indicator_debt = df.groupby('indicator_name')['debt_value'].sum().reset_index().sort_values(by='debt_value', ascending=False)

fig = px.pie(
    indicator_debt.head(5),
    names='indicator_name',
    values='debt_value',
    title='Top 5 Debt Indicators Contributing to Global Debt'
)
fig.show()
```

## 4. Yearly Global Debt Trend

```python
yearly_debt = df.groupby('year')['debt_value'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_debt, x='year', y='debt_value', marker='o', color='crimson')
plt.title('Global Debt Trend Over Time (2010 - 2021)')
plt.xlabel('Year')
plt.ylabel('Total Global Debt (USD)')
plt.grid(True)
plt.show()
```

## 5. Heatmap of Missing Values (Conceptual)

```python
raw_df = pd.read_csv('../data/raw/international_debt.csv')
plt.figure(figsize=(10, 6))
sns.heatmap(raw_df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values in Raw Dataset')
plt.show()
```
